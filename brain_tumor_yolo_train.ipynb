```python
import os
#os.environ['WANDB_MODE']="disabled"
from IPython.display import FileLink
import shutil
from shutil import copyfile
```

```python
# Make a directory to copy the best weights file to
if not os.path.exists('/data1/icmr/shreyasi/brain_tumor_object_detection/output'):
    os.mkdir('/data1/icmr/shreyasi/brain_tumor_object_detection/output')
```

```python
# Specify the series and plane (only the T1wCE series is currently available in the dataset)
SERIES = 't1wce'
PLANE = 'axial'
```

```python
# Call the YOLO train script
!python /data1/icmr/shreyasi/brain_tumor_object_detection/yolov5/train.py --img 512 --batch 32 --epochs 40 --data '/data1/icmr/shreyasi/brain_tumor_object_detection/{PLANE}_{SERIES}_2_class/{PLANE}_{SERIES}_2_class.yaml' --weights /data1/icmr/shreyasi/brain_tumor_object_detection/yolov5/yolov5s.pt --cache
```

```python
# Copy the pt file and zip it up
# Example usage of the exported model: https://www.kaggle.com/davidbroberts/brain-tumor-object-detection

copyfile('/data1/icmr/shreyasi/brain_tumor_object_detection/yolov5/runs/train/exp10/weights/best.pt', f'/data1/icmr/shreyasi/brain_tumor_object_detection/output/{PLANE}_{SERIES}_2_class.pt')
shutil.make_archive(f'{PLANE}_{SERIES}', 'zip', '/data1/icmr/shreyasi/brain_tumor_object_detection/output')

FileLink(f'{PLANE}_{SERIES}.zip')
```
